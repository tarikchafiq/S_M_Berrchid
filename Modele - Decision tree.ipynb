# Import pandas library

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

# Reading the dataset

```python
mydata=pd.read_excel("SM.xlsx")
mydata
```

# Input and output features

```python
X=mydata.drop(['soil moisture'],axis=1)
```

```python
y = mydata['soil moisture']
```

# Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=3)
```

```python
X_test
```

# Decision Tree

```python
from sklearn import tree
model = tree.DecisionTreeRegressor(criterion = 'friedman_mse', max_depth= 9, splitter ='random')
```

```python
%%time
model.fit(X_train, y_train)
```

# Hyper parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(tree.DecisionTreeRegressor(), {
    'splitter': ['random','best'],        
    'max_depth': [ 7, 8, 9, 10, 11, 12, 13, 14, 15],
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson']
}, cv=5, return_train_score=False)
clf.fit(X,y)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_splitter','param_max_depth','param_criterion','mean_test_score']]
```

```python
clf.best_params_
```

# Model score


```python
clf.best_score_
```

# Training model evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
ytrain_pred = model.predict(X_train)
```

```python
print('the training MAE is:%.2f'% mean_absolute_error(y_train,ytrain_pred ))
```

```python
print('the training MSE is:%.2f'% mean_squared_error(y_train,ytrain_pred ))
```

```python
print('the RMSE is:%.2f'% np.sqrt(mean_squared_error(y_train,ytrain_pred )))
```

```python
ytest_pred = model.predict(X_test)
```

# Plotting the observed and predicted data

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(10,6)
x_ax= range(len(X_test))
plt.plot(x_ax, y_test, label = 'Observed', color = 'k', linestyle = '-' )
plt.plot(x_ax, ytest_pred, label = 'Predected', color = 'k', linestyle = '--') 
plt.ylabel('Soil moisture')
plt.xlabel('testing sample data')
plt.legend(bbox_to_anchor=(0.5,-0.2),loc= 'lower center', ncol=2, frameon=False)
```

# Cross validation

```python
from sklearn.model_selection import cross_val_score
score_train= cross_val_score(model, X_train, y_train, scoring = 'neg_mean_squared_error', cv=10)
```

```python
score_train
```

```python
#the absolute mean score
from numpy import absolute
print(absolute(np.mean(score_train)))
```
