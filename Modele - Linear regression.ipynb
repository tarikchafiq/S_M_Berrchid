# Import pandas library

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

# Reading the dataset

```python
mydata=pd.read_excel("SM.xlsx")
mydata
```

# Input and output features

```python
X=mydata.drop(['soil moisture'],axis=1)
```

```python
y = mydata['soil moisture']
```

# Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=3)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression(copy_X= False, fit_intercept= True)
model.fit(X_train,y_train)
```

# Hyper parameter tunning

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(LinearRegression(), {
    'fit_intercept' :[False,True],
     'copy_X' : [False,True]
}, cv=5, return_train_score=False)
clf.fit(X,y)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
clf.best_params_
```

```python
clf.best_score_
```

# Model score


```python
from sklearn.model_selection import cross_val_score
score_train= cross_val_score(model, X_train, y_train, scoring = 'neg_mean_squared_error', cv=10)
```

```python
score_train
```

```python
#the absolute mean score
from numpy import absolute
print(absolute(np.mean(score_train)))
```
